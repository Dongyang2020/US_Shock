## tmp

mean monthly temperature

```python
# create a dataframe with index column
dfindex = pd.read_csv('tmax_1981.csv')
dfindex['ct_proj.atlas_stco'] = dfindex['ct_proj.atlas_stco'].astype(str).str.rjust(5,'0')
dfindex = dfindex[['ct_proj.atlas_stco']]
all_tmp = dfindex.copy()
all_tmp.head()
```

```python
# import required module
import os

# assign directory
directory_tmean = 'tmean_mon'

# iterate over files in
# that directory
arr_tmean = sorted(os.listdir(directory_tmean))
```

```python
arr_tmean
```

```python
# combine all years' April to November data

for i in arr_tmean:
# i=arr[0]
  fname = directory_tmean+"/"+i
  df1 = pd.read_csv(fname,index_col=0)
  yr = df1.loc[1,'year'].astype(str)

  c1 = df1.columns.get_loc("tmean_4")
  c2 = df1.columns.get_loc("tmean_12")
  df1 = df1.iloc[:, c1:c2].reset_index(drop=True)
  all_tmp[yr] = df1.mean(axis = 1)

all_tmp
```

```python
all_tmp.to_csv('tmp.csv')
```

## TX90p

```python
# create a dataframe with index column
dfindex = pd.read_csv('tmax_1981.csv')
dfindex['ct_proj.atlas_stco'] = dfindex['ct_proj.atlas_stco'].astype(str).str.rjust(5,'0')
dfindex = dfindex[['ct_proj.atlas_stco']]
tx90p = dfindex.copy()
all_tmax = dfindex.copy()
all_tmax.head()
```

```python
# import required module
import os

# assign directory
directory_tmax = 'tmax_daily'

# iterate over files in
# that directory
arr_tmax = sorted(os.listdir(directory_tmax))
```

```python
arr_tmax
```

```python
# combine all years' all daily columns

for i in arr_tmax:
# i=arr[0]
  fname = directory_tmax+"/"+i
  df1 = pd.read_csv(fname,index_col=0)
  c1 = df1.columns.get_loc("tmax_0101")
  c2 = df1.columns.get_loc("year")
  df1 = df1.iloc[:, c1:c2].reset_index(drop=True)
  all_tmax = pd.concat([all_tmax, df1], axis=1)

all_tmax
```

```python
all_tmax['p90'] = all_tmax.iloc[:,1::].quantile(
    q=0.9,                      # The percentile to calculate
    axis=1,                     # The axis to calculate the percentile on
    numeric_only=True,          # To calculate only for numeric columns
    interpolation='linear')
all_tmax.head()
```

```python
for i in arr_tmax:
  fname = directory_tmax+"/"+i
  df1 = pd.read_csv(fname,index_col=0)
  yr = df1.loc[1,'year'].astype(str)

  c1 = df1.columns.get_loc("tmax_0401")
  c2 = df1.columns.get_loc("tmax_1201")
  df1 = df1.iloc[:, np.r_[0,c1:c2]].reset_index(drop=True)

  tx90p_count = df1.iloc[:,1::].ge(all_tmax['p90'],axis=0).sum(axis = 1)
  total_count = len(df1.iloc[:,1::].columns)
  tx90p[yr] = tx90p_count/total_count

tx90p.head()
```

```python
tx90p.to_csv('tx90p.csv')
```

## Rx5day
maximum 5-day precipitation intensity

```python
# create a dataframe with index column
dfindex = pd.read_csv('tmax_1981.csv')
dfindex['ct_proj.atlas_stco'] = dfindex['ct_proj.atlas_stco'].astype(str).str.rjust(5,'0')
dfindex = dfindex[['ct_proj.atlas_stco']]

rx = dfindex.copy()
rx.head()
```

```python
# import required module
import os

# assign directory
directory_ppt = 'ppt_daily'


# iterate over files in
# that directory
arr_ppt = sorted(os.listdir(directory_ppt))
```

```python
for i in arr_ppt:

  fname_ppt = directory_ppt+"/"+ i
  df_ppt = pd.read_csv(fname_ppt,index_col=0)
  yr = df_ppt.loc[1,'year'].astype(str)

  c1 = df_ppt.columns.get_loc("ppt_0401")
  c2 = df_ppt.columns.get_loc("ppt_1201")
  df_ppt = df_ppt.iloc[:, c1:c2].reset_index(drop=True)

  rx[yr]=df_ppt.rolling(5,axis=1).sum().max(axis=1)

rx.head()
```

## dtr

```python
# create a dataframe with index column
dfindex = pd.read_csv('tmax_1981.csv')
dfindex['ct_proj.atlas_stco'] = dfindex['ct_proj.atlas_stco'].astype(str).str.rjust(5,'0')
dfindex = dfindex[['ct_proj.atlas_stco']]

dtr = dfindex.copy()
dtr.head()
```

```python
# import required module
import os

# assign directory
directory_tmax = 'tmax_daily'
directory_tmin = 'tmin_daily'

# iterate over files in
# that directory
arr_tmax = sorted(os.listdir(directory_tmax))
arr_tmin = sorted(os.listdir(directory_tmin))
```

```python
# calculate all years' mean range

for i in range(40):
  fname_tmax = directory_tmax+"/"+arr_tmax[i]
  fname_tmin = directory_tmin+"/"+arr_tmin[i]
  df_tmax = pd.read_csv(fname_tmax,index_col=0)
  df_tmin = pd.read_csv(fname_tmin,index_col=0)
  yr = df_tmax.loc[1,'year'].astype(str)
  df_dtr = (df_tmax - df_tmin.values).reset_index(drop=True) # didn't change the columns' label
  df_dtr.head()

  c1 = df_dtr.columns.get_loc("tmax_0401")
  c2 = df_dtr.columns.get_loc("tmax_1201")

  dtr[yr] = df_dtr.iloc[:, c1:c2].mean(axis=1)
dtr.head()
```

```python
dtr.to_csv('dtr.csv')
```

## frs
frost day frequency

```python
# create a dataframe with index column
dfindex = pd.read_csv('tmax_1981.csv')
dfindex['ct_proj.atlas_stco'] = dfindex['ct_proj.atlas_stco'].astype(str).str.rjust(5,'0')
dfindex = dfindex[['ct_proj.atlas_stco']]
frs = dfindex.copy()
frs.head()
```

```python
# import required module
import os

# assign directory
directory_frs = 'tmin_daily'

# iterate over files in
# that directory
arr_frs = sorted(os.listdir(directory_frs))
```

```python
arr_frs
```

```python
for i in arr_frs:
  fname = directory_frs+"/"+i
  df1 = pd.read_csv(fname,index_col=0)
  yr = df1.loc[1,'year'].astype(str)

  c1 = df1.columns.get_loc("tmin_0401")
  c2 = df1.columns.get_loc("tmin_1201")
  df1 = df1.iloc[:, c1:c2].reset_index(drop=True)

  frs_count = df1.iloc[:,1::].lt(0).sum(axis=1)

  frs[yr] = frs_count.div(8)
  # eight months

frs.head()
```

```python
frs.to_csv('frs.csv')
```
